```python
import pandas as pd
import numpy as np
import scipy.stats as stats

path = r"C:\Users\Gustavo\Documents\projetos\detecção fraude\2.python\data\media_ticket_fraudulento_por_hora.csv"
df_agregado = pd.read_csv(path)

valores_tickets = df_agregado['media_valor_fraude']

media_geral = valores_tickets.mean()
desvio_padrao = valores_tickets.std()
n = len(valores_tickets)

# t de Student
limite_inferior, limite_superior = stats.t.interval(
    0.95,
    df=n-1, # graus de liberdade
    loc = media_geral,
    scale = desvio_padrao / np.sqrt(n)
)

print(f"--- PREDICTABLE ANALYSE OF FRAUDULENT TICKET ---")
print(f"Estimated Mean per hour: ${media_geral:.2f}")
print(f"Interval of trust (95%): from ${limite_inferior:.2f} til ${limite_superior:.2f}")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

path = r"C:\Users\Gustavo\Documents\projetos\detecção fraude\2.python\data\transacoes_geral.csv"
df_completo = pd.read_csv(path)

valores_fraude = df_completo[df_completo['classificador'] == 1]['transacoes']

media = valores_fraude.mean()
desvio_padrao = valores_fraude.std()
n = len(valores_fraude)

limite_inferior, limite_superior = stats.norm.interval(
    0.95,
    loc=media,
    scale=desvio_padrao / np.sqrt(n)
)

print(f"Mean: ${media:.2f}")
print(f"Interval of trust (95%): from ${limite_inferior:.2f} til ${limite_superior:.2f}")
```
